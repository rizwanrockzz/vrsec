```python
import keras
from  keras.datasets import mnist
import numpy as NP
import matplotlib.pyplot as plt

```

```python
(X_TRAIN,Y_TRAIN),(X_TEST,Y_TEST) = mnist.load_data()
```

```python
plt.imshow(X_TRAIN[98])
```

```python
plt.imshow(X_TEST[98])
```

```python
X_TRAIN = NP.reshape(X_TRAIN,(60000,784))
X_TEST = NP.reshape(X_TEST,(10000,784))
X_TRAIN = X_TRAIN/255
X_TEST = X_TEST/255
print("Shape of Train Data: {}".format(X_TRAIN.shape))
print("Shape of Train Data: {}".format(X_TEST.shape))
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(32)
pca.fit(X_TRAIN)
```

```python
TRAIN= pca.transform(X_TRAIN)
TEST = pca.transform(X_TEST)
```

```python
TRAIN[3]
```

```python
approximation = pca.inverse_transform(TEST)

```

```python
# Original Image
plt.subplot(1, 2, 1);
plt.imshow(X_TEST[1].reshape(28,28))
#plt.xlabel('784 components', fontsize = 14)
plt.title('Original Image', fontsize = 20);
plt.subplot(1, 2, 2);
plt.imshow(approximation[1].reshape(28, 28))
plt.title('Transformed', fontsize = 20);
```

```python
n = 10  # how many digits we will display
plt.figure(figsize=(20, 4))
for i in range(n):
    # display original
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(X_TEST[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
```

```python
# display reconstruction
n = 10
for i in range(n):
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(approximation[i].reshape(28, 28))
    #plt.imshow(NP.abs(approximation[i]).reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    #plt.show()
```
